```python
%matplotlib inline
import sys
import os
from pathlib import Path
import Orange
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare
from scipy.stats import rankdata
import numpy as np
import pandas as pd

os.chdir(Path(os.getcwd()).resolve().parents[0])
#print(os.getcwd())
import setup
from source import plotFunctions

is_windows = sys.platform.startswith('win')
sep = '\\'

if is_windows == False:
    sep = '/'

path = os.getcwd()+sep+'data'+sep

datasets = [setup.loadCDT, setup.loadCHT, setup.load2CDT, setup.load2CHT, setup.load4CR, setup.load4CRE_V1, 
            setup.load4CRE_V2, setup.load5CVT, setup.loadCSurr, setup.load4CE1CF, setup.loadUG_2C_2D, setup.loadMG_2C_2D, 
            setup.loadFG_2C_2D, setup.loadUG_2C_3D, setup.loadUG_2C_5D, setup.loadGEARS_2C_2D, setup.loadCheckerBoard, 
            setup.loadElecData, setup.loadNOAADataset, setup.loadKeystroke]

RatioToPercent = lambda x,total: (x * 100)/float(total)
CountOcurrence = lambda y,c: (y == c).sum()
timeSteps=[0, 24, 49, 74, 99]

for dataset in datasets:
    dataValues, dataLabels, description = dataset(path, sep)
    
    print(description)
    
    classes = np.unique(dataLabels)
    numberOfClasses = len(classes)
    totalInstances = len(dataValues)
    
    print("Number of instances: {}".format(totalInstances))
    print("Number of atributes: {}".format(dataValues.shape[1]))
    print("Number of classes: {}".format(numberOfClasses))
    for c in classes:
        intancesOfClass = CountOcurrence(dataLabels, c)
        print("Class {:.0f}: {} ({:.2f} %)".format(c, intancesOfClass, RatioToPercent(intancesOfClass, totalInstances)))
        
    try:
        plotFunctions.plotDistributionsTimesteps(dataValues, dataLabels, timeSteps=timeSteps, 
                                                 numberOfBtaches=100, figsize =(15,3))
    except:
        plt.close()
        pass
    
    print("\n")
```

```python
%matplotlib inline
import sys
import os
from pathlib import Path
import Orange
import matplotlib.pyplot as plt
from scipy.stats import friedmanchisquare
from scipy.stats import rankdata
import numpy as np
import pandas as pd

os.chdir(Path(os.getcwd()).resolve().parents[0])
#print(os.getcwd())
import setup
from source import plotFunctions

is_windows = sys.platform.startswith('win')
sep = '\\'

if is_windows == False:
    sep = '/'

path = os.getcwd()+sep+'data'+sep
```

```python
dataValues, dataLabels, description = setup.loadElecData(path, sep)

RatioToPercent = lambda x,total: (x * 100)/float(total)
CountOcurrence = lambda y,c: (y == c).sum()

print(description)

classes = np.unique(dataLabels)
numberOfClasses = len(classes)
totalInstances = len(dataValues)

print("Number of instances: {}".format(totalInstances))
print("Number of atributes: {}".format(dataValues.shape[1]))
print("Number of classes: {}".format(numberOfClasses))
for c in classes:
    intancesOfClass = CountOcurrence(dataLabels, c)
    print("Class {:.0f}: {} ({:.2f} %)".format(c, intancesOfClass, RatioToPercent(intancesOfClass, totalInstances)))


print("\n")

dataValues=pd.DataFrame(dataValues)

display(dataValues.describe())

_= dataValues.hist(figsize=(15,5), layout=(3,2))
```
